## Задание 1
Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

если источник traffic_source равен yandex или google, то в source_type ставится organic
для источников paid и email из России - ставим ad
для источников paid и email не из России - ставим other
все остальные варианты берем из traffic_source без изменений

```python
import pandas as pd
```

```python
log = pd.read_csv('visit_log.csv', sep=';')
log.head(10)
```

```python
log.loc[
    log['traffic_source'].isin(['yandex', "google"]),'traffic_type'] = 'organic'
    
log.head(15)
```

```python
log.loc[
    (log.region == 'Russia') & 
    (log.traffic_source.isin(['email', "paid"])),
    'traffic_type'
] = 'ad' 
```

```python
log.head(15)
```

```python
# проверила - что не Россия не попали под этот код
log.loc[log.traffic_source== 'email'].head(15)
```

```python
log.loc[
    (log.region != 'Russia') & 
    (log.traffic_source.isin(['email', "paid"])),
    'traffic_type'
] = 'other' 
```

```python
log.loc[log.traffic_source== 'paid'].head(15)
```

```python
log.loc[pd.isnull(log.traffic_type), 'traffic_type'] = log.traffic_source
```

```python
log.head(17)
```

## Домашнее задание 3
В датафрейме data создайте столбец lemmas, 
в котором вычислите леммы поисковых запросов из столбца keyword. Леммы должны иметь строковый тип.

```python
from pymystem3 import Mystem
```

```python
data = pd.DataFrame({
    'keyword': ['курс гривны к рублю', 'доллары в рубли', '100 долларов в рублях', 'курс рубля'],
    'shows': [125076, 114173, 97534, 53546],
})
```

```python
m= Mystem()
```

```python
data['lemmas'] = data.keyword.apply(m.lemmatize)
```

```python
data
```

```python
data['lemmas'] = data.lemmas.apply(''.join)
```

```python
data
```

```python
def list_str(lemmas):
    return lemmas.strip()
```

```python
data.lemmas = data['lemmas'].apply(list_str)
```
